```python
# Combine three datasets:
import numpy as np
import pandas as pd

data1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")

data2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")

data3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
data1.info()
```

```python
data2.head(2)
```

```python
data3.head(3)
```

```python
data_comb= pd.concat([data1, data2, data3], axis=0).head(5)
data_comb.info()
```

```python
# cleaning
data_comb.columns = data_comb.columns.str.lower() 

data_comb.columns = data_comb.columns.str.replace(' ', '_')

data_comb.rename(columns={'st': 'state'})

data_comb['gender']= data_comb['gender'].replace({'Femal':'F', 'female':'F', 'Femal': 'F', 'Male':'M'})

data_comb['state']= data_comb['state'].replace({'Cali': 'California', 'Wa': 'Washington','AZ': 'Arizona'})

data_comb['education']= data_comb['education'].replace({'Bachelors':'Bachelor'})

data_comb['customer_lifetime_value'] = data_comb['customer_lifetime_value'].str.replace('%', '') 

data_comb['vehicle_class'] = data_comb['vehicle_class'].replace(
    {'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'})
```

```python
data_comb.head(5)
```

```python
#formatting data types

data_comb['customer_lifetime_value'] = data_comb['customer_lifetime_value'].astype(float)

data_comb['number_of_open_complaints'] = data_comb['number_of_open_complaints'].astype(str).str.split('/').str[1]

data_comb['number_of_open_complaints'] = pd.to_numeric(data_comb['number_of_open_complaints'])
```

```python
data_comb.head(2)
```

```python
#dealing with null values

cl_data_comb= data_comb.dropna(how='all')

cl_data_comb['gender'] = cl_data_comb['gender'].fillna(cl_data_comb['gender'].mode())

cl_data_comb['customer_lifetime_value']= cl_data_comb['customer_lifetime_value'].fillna(cl_data_comb['customer_lifetime_value'].median())
```

```python
cl_data_comb.isna().any()
```

```python
# dealing with duplicates

cl_data_comb = cl_data_comb.drop_duplicates(subset='customer', keep='first')
cl_data_comb
```

```python
cl_data_comb.columns
```
